```python
import requests
import torch
import torch.nn as nn
import torch.optim as optim
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
from transformers import BertTokenizer, BertForSequenceClassification
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Certifique-se de baixar os pacotes necessários
import nltk
nltk.download('punkt')
nltk.download('stopwords')

# URL do arquivo RAW
url_arquivo = 'https://raw.githubusercontent.com/JotaChina/Computer-intelligence/update%231/TP2/RacionaisLyrics.raw'

# Fazer solicitação HTTP para obter o conteúdo do arquivo
response = requests.get(url_arquivo)
if response.status_code == 200:
    texto_raw = response.text
else:
    print("Falha ao fazer solicitação HTTP")

# Pré-processamento do texto
def preprocess_text(text):
    tokens = word_tokenize(text, language='portuguese')
    stop_words = set(stopwords.words('portuguese'))
    stemmer = RSLPStemmer()
    preprocessed_tokens = [stemmer.stem(token) for token in tokens if token.lower() not in stop_words]
    preprocessed_text = ' '.join(preprocessed_tokens)
    return preprocessed_text

# Dividir texto em linhas
lines = texto_raw.split('\n')

# Pré-processar linhas
preprocessed_lines = [preprocess_text(line) for line in lines if line.strip()]

# Criando pares de entrada e saída
seq_length = 3  # Número de linhas na sequência de entrada
data = []

for i in range(len(preprocessed_lines) - seq_length):
    input_seq = preprocessed_lines[i:i+seq_length]
    target_seq = preprocessed_lines[i+seq_length]
    data.append((input_seq, target_seq))

# Tokenização
tokenizer = Tokenizer()
tokenizer.fit_on_texts(preprocessed_lines)

# Definindo comprimento máximo
max_len_input = seq_length  # Porque estamos usando 3 linhas de entrada
max_len_output = 1  # Porque estamos prevendo a próxima linha

# Função para codificar e padronizar sequências
def encode_and_pad_sequences(sequences, max_len):
    encoded_sequences = tokenizer.texts_to_sequences(sequences)
    padded_sequences = pad_sequences(encoded_sequences, maxlen=max_len, padding='post')
    return padded_sequences

# Preparando os dados para o treinamento
encoded_data = [(encode_and_pad_sequences(input_seq, max_len_input),
                 encode_and_pad_sequences([target_seq], max_len_output)) for input_seq, target_seq in data]

# Convertendo para tensores
inputs = torch.tensor([input_seq for input_seq, _ in encoded_data], dtype=torch.long)
targets = torch.tensor([target_seq[0] for _, target_seq in encoded_data], dtype=torch.long)

# Ajustando a forma das entradas para 3D
inputs = inputs.view(-1, max_len_input)  # Transformar em 2D antes de dataloader
targets = targets.view(-1)  # Transformar em 1D

# Criando DataLoader
from torch.utils.data import DataLoader, TensorDataset

dataset = TensorDataset(inputs, targets)
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```
